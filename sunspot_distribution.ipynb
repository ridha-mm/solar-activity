```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
from astropy.io import ascii
```

```python
data = ascii.read('sunspots-data')
```

```python
plt.hist(data['lat'], bins=18, color='dimgray')
plt.xlabel('Latitude (degrees)')
plt.ylabel('Number of sunspots')
```

```python
plt.scatter(data['Year'], data['lat'], color='dimgray', alpha=0.1 , s=10, edgecolors='None')
plt.xlabel('Year')
plt.ylabel('Latitude (degrees)')
```

```python
plt.scatter(data['Year'], data['long'], color='dimgray', alpha=0.1 , s=10, edgecolors='None')
plt.xlabel('Year')
plt.ylabel('Carrington longitude (degrees)')
```
